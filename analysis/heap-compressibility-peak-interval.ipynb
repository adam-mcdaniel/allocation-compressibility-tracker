```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# The data is stored in a 5XX.csv for the serial tests, and 6XX.csv for the parallel tests
# The data files are in the `/data` folder
DIRECTORY = "../data/raw-data"


if not 'all_data' in globals() and not 'all_data' in locals():
    # A structure for storing the data
    all_data = {}

    for i in range(500, 600):
        file = f"{DIRECTORY}/{i}/object-compression.csv"
        if os.path.isfile(file):
            all_data[i] = pd.read_csv(file)
            print(f"Loaded data for test {i}")
```

```python
# For each test, try to use the RandomForestClassifier to predict the "Compression Class" using the "Allocation Site" column
# Only looking at one "Compression Type" at a time

# Sort the test numbers by the number of rows in the data,
# from smallest to largest
# sorted_test_numbers = sorted(all_data.keys(), key=lambda x: len(all_data[x]))
sorted_test_numbers = sorted(all_data.keys(), key=lambda x: x)
print(sorted_test_numbers)
# Create a dataframe to store all the feature importances for each test
compression_results = pd.DataFrame()

# Create a dataframe to store all the peak RSS data for each test
peak_rss_results = None
print(f"Found {len(sorted_test_numbers)} tests")
for test_number in sorted_test_numbers:
    # The data for the test
    test_data = all_data[test_number]
    
    # Get the unique values for the "Compression Type" column
    compression_types = test_data["Compression Type"].unique()
    print(f"Test {test_number} has {len(compression_types)} compression types: {compression_types}")
    
    # # For each compression type, try to predict the "Compression Class" using the "Allocation Site" column
    # for compression_type in compression_types:
    
    
    try:
        print(test_data)
        # Get the interval with the peak resident set size
        # Calculate this by summing the object sizes for each interval
        # and taking the maximum
        test_data["Peak RSS"] = test_data.groupby("Interval #")["Size (bytes)"].transform("sum")
        
        # Find the interval with the peak RSS
        peak_rss_interval = test_data.loc[test_data["Peak RSS"].idxmax()]["Interval #"]
        
        # Get the data for the interval with the peak RSS
        peak_rss_data = test_data[test_data["Interval #"] == peak_rss_interval]
        
        # # Get the unique values for the "Compression Class" column
        # compression_classes = peak_rss_data["Compression Class"].unique()
        
        # Get the average compressibility of the heap
        # by calculating the average "Compression Ratio (compressed/uncompressed)" column
        average_compressibility = peak_rss_data["Compression Ratio (compressed/uncompressed)"].mean()
        
        # Get the average size of the heap
        # by calculating the average "Size (bytes)" column
        average_object_size = peak_rss_data["Size (bytes)"].mean()
        
        # Show a violin plot of the compressibility
        # and the object size
        print(peak_rss_data)
        # fig, ax = plt.subplots(1, 2, figsize=(10, 5))
        # fig, ax = plt.subplots(1, 1, figsize=(8, 8))
        fig, ax = plt.subplots(1, 1)
        sns.violinplot(data=peak_rss_data, y="Size (bytes)", ax=ax, label="Distribution Of Object Sizes", cut=0)
        sns.violinplot(data=peak_rss_data, y="Compressed Size (bytes)", ax=ax, label="Distribution Of Compressed Object Sizes", cut=0)
        # sns.violinplot(data=peak_rss_data, y="Size (bytes)", ax=ax[1])
        plt.suptitle(f"Test {test_number} - Peak RSS Interval {peak_rss_interval} - Average Compressibility {average_compressibility:.2f} - Average Object Size {average_object_size:.2f}")
        plt.show()
        # tmp = {
        #     "Test Number": test_number,
        #     "Peak RSS Interval": peak_rss_interval,
        #     "Average Compressibility": average_compressibility,
        #     "Average Object Size": average_object_size,
        #     # Include the peak RSS data
        #     "Peak RSS Data": peak_rss_data
        # }
        peak_rss_data["Test Number"] = f'{test_number}\n({peak_rss_interval * 15}s)'
        peak_rss_data["Peak RSS Interval"] = peak_rss_interval
        peak_rss_data["Average Compressibility"] = average_compressibility
        peak_rss_data["Average Object Size"] = average_object_size
        # Set the test number for all rows
        # Add the columns to the dataframe
        
        # peak_rss_data.assign(Test_Number=test_number)
        # peak_rss_data.assign(Peak_RSS_Interval=peak_rss_interval)
        # peak_rss_data.assign(Average_Compressibility=average_compressibility)
        # peak_rss_data.assign(Average_Object_Size=average_object_size)
        print(peak_rss_data.columns)
        if peak_rss_results is None:
            peak_rss_results = peak_rss_data
        else:
            peak_rss_results = pd.concat([peak_rss_results, peak_rss_data], ignore_index=True)
        # print(tmp)
        # tmp = pd.DataFrame(tmp, index=[0])
        # print("TMP", tmp)
    except Exception as e:
        # Print as red
        print(f'\033[91mFailed to evaluate Test {test_number}: {e}\033[0m')
        raise e
```

```python
# Graph all the results at once from the peak_rss_results dataframe
print(peak_rss_results.columns)


# Create a dataframe to store the sizes of the objects, compressed and uncompressed.
# There are 2 columns: Size, Is Compressed?
size_results = pd.DataFrame()
# Add all the uncompressed sizes with the label "Uncompressed"
size_results = pd.concat([size_results, peak_rss_results[["Size (bytes)", "Test Number", "Compression Ratio (compressed/uncompressed)"]].rename(columns={"Size (bytes)": "Size", "Test Number": "Test Number", "Peak RSS Interval": "Peak RSS Interval", "Average Compressibility": "Average Compressibility", "Average Object Size": "Average Object Size"})], ignore_index=True)
# Set is_compressed to False
size_results["Is Compressed"] = "Uncompressed"

# Add all the compressed sizes with the label "Compressed"
compressed_subset = peak_rss_results[["Compressed Size (bytes)", "Test Number", "Compression Ratio (compressed/uncompressed)"]].rename(columns={"Compressed Size (bytes)": "Size", "Test Number": "Test Number", "Peak RSS Interval": "Peak RSS Interval", "Average Compressibility": "Average Compressibility", "Average Object Size": "Average Object Size"})
compressed_subset["Is Compressed"] = "Compressed"
size_results = pd.concat([size_results, compressed_subset], ignore_index=True)

# Graph the violin plots for all the test numbers
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
print(size_results.columns)
sns.violinplot(data=size_results, x="Test Number", y="Size", hue="Is Compressed", ax=ax, cut=0, inner="quart", density_norm='width', split=True)
plt.suptitle(f"Heap Compressibility During Interval Of Peak RSS For CPU2017")
plt.xlabel("\nTest Number\n(When Peak RSS Was Reached)")
plt.show()
# Show the same graph, but only for objects < 5 * 10e7 bytes
fig, ax = plt.subplots(1, 1, figsize=(16, 4))
sns.violinplot(data=size_results, x="Test Number", y="Compression Ratio (compressed/uncompressed)", ax=ax, cut=0, inner="quart", density_norm='width')
plt.xlabel("\nTest Number\n(When Peak RSS Was Reached)")
# Set the y-axis to be until 5e7
# plt.ylim(0, 5e8)
plt.suptitle(f"Heap Compressibility During Interval Of Peak RSS For CPU2017 (Distribution Of Object Compression Ratios)")

plt.show()
```
