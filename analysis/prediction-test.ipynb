```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# The data is stored in a 5XX.csv for the serial tests, and 6XX.csv for the parallel tests
# The data files are in the `/data` folder
DIRECTORY = "../data/prediction-test"


if not 'all_data' in globals() and not 'all_data' in locals():
    # A structure for storing the data
    all_data = {}

    for i in range(500, 600):
        if os.path.isfile(f"{DIRECTORY}/{i}.csv"):
            all_data[i] = pd.read_csv(f"{DIRECTORY}/{i}.csv")
            print(f"Loaded data for test {i}")
```

```python
# For each test, try to use the RandomForestClassifier to predict the "Compression Class" using the "Allocation Site" column
# Only looking at one "Compression Type" at a time

# Sort the test numbers by the number of rows in the data,
# from smallest to largest
sorted_test_numbers = sorted(all_data.keys(), key=lambda x: len(all_data[x]))

# Create a dataframe to store all the feature importances for each test
compression_results = pd.DataFrame()

print(f"Found {len(sorted_test_numbers)} tests")
for test_number in sorted_test_numbers:
    # The data for the test
    test_data = all_data[test_number]
    
    # Get the unique values for the "Compression Type" column
    compression_types = test_data["Compression Type"].unique()
    print(f"Test {test_number} has {len(compression_types)} compression types: {compression_types}")
    
    # For each compression type, try to predict the "Compression Class" using the "Allocation Site" column
    for compression_type in compression_types:
        try:
            if isinstance(compression_type, float) and np.isnan(compression_type):
                continue
            print(f"Test {test_number} Compression Type {compression_type}")
            
            # Get the data for the compression type
            compression_data = test_data[test_data["Compression Type"] == compression_type]
            
            # Get the unique values for the "Allocation Site" column
            allocation_sites = compression_data["Allocation Site"].unique()
            print(f"Test {test_number} Compression Type {compression_type} has {len(allocation_sites)} allocation sites: {allocation_sites}")
            
            # Create a dictionary to store the results
            results = {}
            
            # Try to predict the compression classes using only the allocation sites, age, and size
            # allocation_data = compression_data[["Allocation Site", "Size (bytes)"]]
            # allocation_data = compression_data[["Size (bytes)"]]
            allocation_data = compression_data[["Allocation Site", "Age (intervals)", "Size (bytes)"]]
            # Convert allocation site to a one-hot encoding
            allocation_data = pd.get_dummies(allocation_data, columns=["Allocation Site"])
            
            X = allocation_data
            y = compression_data["Compression Class"]
            
            # Split the data into training and testing sets
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
            print(f"Test {test_number} Compression Type {compression_type} has {len(X_train)} training samples and {len(X_test)} testing samples")
            # Create the RandomForestClassifier
            # rfc = RandomForestClassifier(n_estimators=100)
            # Use only as many trees as there are features
            rfc = RandomForestClassifier(n_estimators=X.shape[1])
            print(f"Created RandomForestClassifier with {rfc.n_estimators} trees")
            
            rfc.fit(X_train, y_train)
            
            # Predict the compression classes
            predictions = rfc.predict(X_test)
            
            # Store the classification report
            results["compression"] = classification_report(y_test, predictions)
            
            # Store the confusion matrix
            results["compression_matrix"] = confusion_matrix(y_test, predictions)
            # Show both the classification report and confusion matrix
            print(compression_type, results["compression"])
            print(compression_type, results["compression_matrix"])
            
            # Show the feature importances
            feature_importances = pd.DataFrame(rfc.feature_importances_, index=X.columns, columns=["Importance"]).sort_values("Importance", ascending=False)
            # Combine all the one-hot columns into a single column
            feature_importances = feature_importances.groupby(feature_importances.index.str.split("_").str[0]).sum()
            feature_importances = feature_importances.sort_values("Importance", ascending=False)
            
            
            print(compression_type, feature_importances)
            # Show a bar chart of the feature importances
            feature_importances.plot(kind="bar")
            plt.title(f'Test {test_number} Compression Type {compression_type} Feature Importances')
            plt.show()
            
            # Show as a seaborn heatmap
            plt.title(f'Test {test_number} Compression Type {compression_type} Prediction')
            sns.heatmap(results["compression_matrix"], annot=True)
            plt.show()
            
            class_predictions = classification_report(y_test, predictions, output_dict=True)
            print(class_predictions)
            # Get average precision
            average_precision = class_predictions['weighted avg']['precision']
            
            
            # Store the feature importances, and the accuracy / precision / recall / f1-score for each test
            # Each row is a different test
            # Each column is a different feature
            # The last column is the overall percentage correct
            compression_output = {
                "test": test_number,
                "compression_type": compression_type,
                # Now add the feature importances
                **{f"Importance of {name}": importance for name, importance in feature_importances["Importance"].items()},
                "accuracy": class_predictions["accuracy"],
                "precision": class_predictions["weighted avg"]["precision"],
                "recall": class_predictions["weighted avg"]["recall"],
                "f1-score": class_predictions["weighted avg"]["f1-score"],
                "overall_percentage_correct": np.sum(np.diag(results["compression_matrix"])) / np.sum(results["compression_matrix"]) * 100,
            }
            print('to_dict: ', compression_output)
            compression_results = pd.concat([compression_results, pd.DataFrame([compression_output], columns = compression_output.keys())], ignore_index=True)
            
            # Save the results to a CSV file
            compression_results.to_csv("compression_results.csv")
            # Print overall percentage correct
            print(f"Test {test_number}  Compression Type {compression_type} Overall percentage correct: {np.sum(np.diag(results['compression_matrix'])) / np.sum(results['compression_matrix']) * 100:.2f}%")
        except Exception as e:
            # Print as red
            print(f'\033[91mFailed to evaluate Test {test_number} Compression Type {compression_type}: {e}\033[0m')
```

```python
# Go through the `evaluation_results` dataframe and graph the feature importances and the overall percentage correct for each test
evaluation_results = pd.read_csv("compression_results.csv")

# Get the compression types
compression_types = evaluation_results["compression_type"].unique()

# For each compression type, graph the feature importances and the overall percentage correct for each test
for compression_type in compression_types:
    # Get the data for the compression type
    compression_type_data = evaluation_results[evaluation_results["compression_type"] == compression_type]
    
    # Get the test numbers
    test_numbers = compression_type_data["test"].unique()
    
    plt.figure(figsize=(30, 5))
    fig, ax1 = plt.subplots()
    # Set the width and height of the figure
    fig.set_figwidth(10)
    fig.set_figheight(5)

    # Set the y axis to be a percentage
    plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter())
    feature_importances_columns = ['Importance of Size (bytes)', 'Importance of Age (intervals)', 'Importance of Allocation Site']
    short_feature_importances_columns = ['Size', 'Age', 'Site']
    # For each test, graph the feature importances and the overall percentage correct as a bar chart
    # Do a unique color for each test
    bars = []
    
    test_numbers = sorted(list(test_numbers))
    
    for test_number in test_numbers:
        test_row = compression_type_data[compression_type_data["test"] == test_number].to_dict(orient="records")[0]
        
        print(test_row)
        feature_importances = [test_row[name] * 100 for name in feature_importances_columns]
        print('Feature importances:', feature_importances)
        overall_percentage_correct = test_row["overall_percentage_correct"]

        # keys = [f"#{test_number} {name}" for name in short_feature_importances_columns] + [f"#{test_number} Overall Percentage Correct"]
        # values = [*feature_importances, overall_percentage_correct]
        keys = [f"#{test_number} {name}" for name in short_feature_importances_columns]
        values = [*feature_importances]
        
        bars.append([keys, values])
        
    # Now show the average across all tests in a different color
    average_feature_importances = [value * 100 for value in compression_type_data[feature_importances_columns].mean().values]
    average_overall_percentage_correct = compression_type_data["overall_percentage_correct"].mean()
    # average_keys = [f"Average {name}" for name in short_feature_importances_columns] + ["Average Overall Percentage Correct"]
    # average_values = [*average_feature_importances, average_overall_percentage_correct]
    average_keys = [f"Average {name}" for name in short_feature_importances_columns]
    average_values = [*average_feature_importances]
    bars.append([average_keys, average_values])
    
    colors = plt.cm.viridis(np.linspace(0, 1, len(test_numbers) + 1))
    all_keys = []
    
    # Plot each set of data
    for i, data in enumerate(bars):
        color = colors[i]
        bar_width = 1.5  # Width of the bars
        # Add gaps between each set of bars
        keys = data[0]
        values = data[1]
        offsets = bar_width * np.arange(start=i * len(keys), stop=i * len(keys) + len(keys))
        print('Keys:', keys)
        print('Values:', values)
        all_keys += keys
        # Create bar plot
        bars = ax1.bar(offsets, values, bar_width, color=color, alpha=1.0, label=f'Test {test_numbers[i]}' if i < len(test_numbers) else 'Average')
        
        
    # Set the names for the x-axis
    ax1.set_xticks(np.arange(len(all_keys)) * bar_width)
    ax1.set_xticklabels(all_keys)
    # Make the x-axis labels vertical
    plt.yticks(np.arange(0, 101, 5))
    plt.xticks(rotation=90)

    for key, value in zip(average_keys, average_values):
        plt.axhline(y=value, color='red', linestyle='--')
        plt.text(x=40, y=value+0.1, s=f'{key} importance', ha='right', va='bottom', color='r')

    
    plt.title(f"Compression Type {compression_type} Feature Importances")
    # plt.legend()
    plt.show()
    
    # Now show the overall percentage correct
    plt.figure(figsize=(10, 5))
    # Count the compression classes
    compression_classes = all_data[test_number]["Compression Class"].value_counts()
    plt.title(f"Compression Type {compression_type} Overall Percentage Correct (Predicting From {len(compression_classes)} Classes)")
    plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter())
    # Set number of y-ticks
    plt.yticks(np.arange(0, 101, 5))
    plt.bar([f"Test {test_number}" for test_number in test_numbers], compression_type_data["overall_percentage_correct"], color='blue', alpha=0.5, label='Test')
    plt.bar("Average", average_overall_percentage_correct, color='red', alpha=0.5, label='Average')
    # Draw a flat line at the average
    plt.axhline(y=average_overall_percentage_correct, color='red', linestyle='--')
    plt.legend()
    plt.show()
```

```python
# For each test, try to use the RandomForestClassifier to predict the "Compression Class" using the "Allocation Site" column
# Only looking at one "Compression Type" at a time

# Sort the test numbers by the number of rows in the data,
# from smallest to largest
sorted_test_numbers = sorted(all_data.keys(), key=lambda x: len(all_data[x]))

# Create a dataframe to store all the feature importances for each test
access_results = pd.DataFrame()

print(f"Found {len(sorted_test_numbers)} tests")
for test_number in sorted_test_numbers:
    # The data for the test
    test_data = all_data[test_number]
    
    # Get the unique values for the "Compression Type" column
    compression_types = test_data["Compression Type"].unique()
    print(f"Test {test_number} has {len(compression_types)} compression types: {compression_types}")
    try:
        print(f"Test {test_number}")
        
        # Get the unique values for the "Allocation Site" column
        allocation_sites = test_data["Allocation Site"].unique()
        print(f"Test {test_number} has {len(allocation_sites)} allocation sites: {allocation_sites}")
        
        # Create a dictionary to store the results
        results = {}
        
        # Try to predict the compression classes using only the allocation sites, age, and size
        # allocation_data = compression_data[["Allocation Site", "Size (bytes)"]]
        # allocation_data = compression_data[["Size (bytes)"]]
        allocation_data = test_data[["Allocation Site", "Age (intervals)", "Size (bytes)"]]
        # Convert allocation site to a one-hot encoding
        allocation_data = pd.get_dummies(allocation_data, columns=["Allocation Site"])
        
        X = allocation_data
        y = test_data["Access Type"]
        
        # Split the data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
        print(f"Test {test_number} has {len(X_train)} training samples and {len(X_test)} testing samples")
        # Create the RandomForestClassifier
        # rfc = RandomForestClassifier(n_estimators=100)
        # Use only as many trees as there are features
        rfc = RandomForestClassifier(n_estimators=X.shape[1])
        print(f"Created RandomForestClassifier with {rfc.n_estimators} trees")
        
        rfc.fit(X_train, y_train)
        
        # Predict the compression classes
        predictions = rfc.predict(X_test)
        
        # Store the classification report
        results["access"] = classification_report(y_test, predictions)
        
        # Store the confusion matrix
        results["access_matrix"] = confusion_matrix(y_test, predictions)
        # Show both the classification report and confusion matrix
        print(test_number, results["access"])
        print(test_number, results["access_matrix"])
        
        # Show the feature importances
        feature_importances = pd.DataFrame(rfc.feature_importances_, index=X.columns, columns=["Importance"]).sort_values("Importance", ascending=False)
        # Combine all the one-hot columns into a single column
        feature_importances = feature_importances.groupby(feature_importances.index.str.split("_").str[0]).sum()
        feature_importances = feature_importances.sort_values("Importance", ascending=False)
        
        
        print(feature_importances)
        # Show a bar chart of the feature importances
        feature_importances.plot(kind="bar")
        plt.title(f'Test {test_number} Feature Importances')
        plt.show()
        
        # Show as a seaborn heatmap
        plt.title(f'Test {test_number} Prediction')
        sns.heatmap(results["access_matrix"], annot=True)
        plt.show()
        
        access_predictions = classification_report(y_test, predictions, output_dict=True)
        print(access_predictions)

        access_output = {
            "test": test_number,
            **{f"Importance of {name}": importance for name, importance in feature_importances["Importance"].items()},
            "accuracy": access_predictions["accuracy"],
            "precision": access_predictions["weighted avg"]["precision"],
            "recall": access_predictions["weighted avg"]["recall"],
            "f1-score": access_predictions["weighted avg"]["f1-score"],
            "overall_percentage_correct": np.sum(np.diag(results["access_matrix"])) / np.sum(results["access_matrix"]) * 100,
        }
        print('to_dict: ', access_output)
        access_results = pd.concat([access_results, pd.DataFrame([access_output], columns = access_output.keys())], ignore_index=True)
        
        # Save the results to a CSV file
        access_results.to_csv("access_results.csv")
        # Print overall percentage correct
        print(f"Test {test_number}  Overall percentage correct: {np.sum(np.diag(results['access_matrix'])) / np.sum(results['access_matrix']) * 100:.2f}%")
    except Exception as e:
        # Print as red
        print(f'\033[91mFailed to evaluate Test {test_number}: {e}\033[0m')
```
